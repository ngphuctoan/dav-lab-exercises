```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from setuptools.command.rotate import rotate
from scipy import stats

plt.style.use("ggplot")
plt.rcParams.update({
    "font.size": 12,
})
```

```python
statcounter_ww = pd.read_csv("../data/linux-adoption/StatCounter_Worldwide.csv", parse_dates=["Date"])
statcounter_vn = pd.read_csv("../data/linux-adoption/StatCounter_Viet+Nam.csv", parse_dates=["Date"])

steam_hw_survey = pd.read_csv("../data/linux-adoption/Steam+HW+Survey_Platform+Versions.csv", parse_dates=["date"])
```

```python
statcounter_merged = statcounter_ww.merge(
    statcounter_vn,
    on="Date",
    suffixes=("_WW", "_VN"),
)

statcounter_merged[["Linux_WW", "Linux_VN"]].describe()
```

```python
drop_percentage = 10
confidence_interval = 0.95

win_latest = statcounter_ww["Windows"].iloc[-1]
linux_latest = statcounter_ww["Linux"].iloc[-1]
others_latest = 100 - win_latest - linux_latest

E_win = win_latest - drop_percentage
E_linux = 100 - E_win - others_latest

linux_std = (statcounter_ww["Linux"].pct_change() * 100).std()
z = stats.norm.ppf(confidence_interval)

ci_lower = E_linux - z * linux_std
ci_upper = E_linux + z * linux_std

print(f"E[Linux | Windows ↓{drop_percentage}%] = {E_linux:.2f}%")
print(f"{confidence_interval * 100}% CI: [{ci_lower:.2f}%, {ci_upper:.2f}%]")
```
