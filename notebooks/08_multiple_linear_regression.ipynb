### Multiple Linear Regression

Proposed question:

*"Which laptop specifications significantly affect the price, and how?"*

**Explanation:**

We perform an Ordinary Least Squares (OLS) regression using CPU frequency, screen size, weight, RAM, storage, and screen resolution to predict price. This quantifies each feature's contribution while controlling for the others.

```python
import statsmodels.api as sm

X = df[["CPU_freq", "Inches", "Weight", "Ram", "PrimaryStorage", "ScreenW", "ScreenH"]]
y = df["Price_euros"]

X = sm.add_constant(X)

model = sm.OLS(y, X).fit()

model.summary()
```

Summary:

* CPU frequency and RAM are strong positive drivers of price.
* Larger screens slightly decrease price; heavier laptops slightly increase it.
* Storage has a small negative effect; screen resolution (width & height) is mostly insignificant.

Overall, the model explains ~67% of price variance ($R^2 = 0.688$), confirming that processors and RAM dominate price determination in practice.
